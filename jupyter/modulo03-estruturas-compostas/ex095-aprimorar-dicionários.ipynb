Exercício Python 095: Aprimore o desafio 93 para que ele funcione com vários jogadores, incluindo um sistema de visualização de detalhes do aproveitamento de cada jogador.

```python
jogador = dict()
partidas = list()

jogador['nome'] = str(input('Nome do Jogador: '))
total_partidas = int(input(f'Quantas partidas {jogador["nome"]} jogou? '))
                           
for elemento in range(0, total_partidas):
    partidas.append(int(input(f'Quantos gols na partida {elemento + 1}? ')))
                           
jogador['gols'] = partidas[:]
jogador['total_gols'] = sum(partidas)
                           
print('-' * 30)
print(jogador)
print('-' * 30)

for chave, valor in jogador.items():
    print(f'O campo {chave} tem o valor {valor}')
    
print('-' * 30)
print(f'O jogador {jogador["nome"]} jogou {len(jogador["gols"])} partidas.')


for indice, valor in enumerate(jogador['gols']):
    print(f'    => Na partida {indice + 1} fez {valor} gols')
    
print(f'Fez um total de {jogador["total_gols"]} gols.')
```

```python
equipe = list()
jogador = dict()
partidas = list()

while True:
    jogador.clear()
    jogador['nome'] = str(input('Nome do Jogador: '))
    total_partidas = int(input(f'Quantas partidas {jogador["nome"]} jogou? '))
    partidas.clear()

    for elemento in range(0, total_partidas):
        partidas.append(int(input(f'    Quantos gols na partida {elemento + 1}? ')))

    jogador['gols'] = partidas[:] # para nao incluir todas as partidas de todos os jogadores
    jogador['total_gols'] = sum(partidas)
    
    equipe.append(jogador.copy())
          
    while True:
        resposta = str(input('Quer continuar? [S/N] ')).upper()[0]
        if resposta in 'SN':
            break
        print('ERRO! Responda somente S ou N.')
    if resposta == 'N':
        break
    

print('-' * 40)
print('cod ', end='')
for indice in jogador.keys():
    print(f'{indice:<15}', end='')
print()
    
print('-' * 40)
for chave, valor in enumerate(equipe):
    print(f'{chave:>3} ', end='')
    for dado in valor.values():
        print(f'{str(dado):<15}', end='')
    print()
print('-' * 40)


while True:
    busca = int(input('Mostrar os dados de qual jogador? (999 interrompe) '))
    if busca == 999:
        break
    if busca >= len(equipe):
        print(f'ERRO! Não existe jogador com o código {chave}.')
    else:
        print(f' -- LEVANTAMENTO DO JOGADOR {equipe[busca]["nome"]}')
        for indice, gols in enumerate(equipe[busca]['gols']):
            print(f'    No jogo {indice + 1} fez {gols} gols.')

```
